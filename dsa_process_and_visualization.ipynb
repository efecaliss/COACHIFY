```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Dosyayı oku
df = pd.read_excel("/content/DSA_Student_List_Final.xlsx", sheet_name="Merged")

# Scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(df["YKS_Ranking"], df["Pass_Rate"], color='blue')
plt.xlabel("YKS Ranking")
plt.ylabel("Pass Rate")
plt.title("YKS Ranking vs. Pass Rate")
plt.grid(True)
plt.show()

# Pearson correlation
corr, pval = pearsonr(df["YKS_Ranking"], df["Pass_Rate"])
print("Pearson correlation:", round(corr, 3))
print("p-value:", round(pval, 4))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind

# Dosyayı oku
df = pd.read_excel("/content/DSA_Student_List_Final.xlsx", sheet_name="Merged")

# Boxplot
sns.boxplot(x="Primary_Communication_Format", y="Pass_Rate", data=df)
plt.xlabel("Primary Communication Format (video / voice)")
plt.ylabel("Pass Rate")
plt.title("Pass Rate by Communication Format")
plt.grid(True)
plt.show()

# t-test
video = df[df["Primary_Communication_Format"] == "video"]["Pass_Rate"]
voice = df[df["Primary_Communication_Format"] == "voice"]["Pass_Rate"]

t_stat, p_val = ttest_ind(video, voice, equal_var=False)

print("t-statistic:", round(t_stat, 3))
print("p-value:", round(p_val, 4))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Dosyayı oku
df = pd.read_excel("/content/DSA_Student_List_Final.xlsx", sheet_name="Merged")

# Bar plot: Monthly Progress Messages gruplandırarak
df_grouped = df.groupby("Weekly_Progress_Messages")["Pass_Rate"].mean().reset_index()

plt.figure(figsize=(8, 5))
plt.bar(df_grouped["Weekly_Progress_Messages"], df_grouped["Pass_Rate"], color='purple')
plt.xlabel("Monthly Progress Messages")
plt.ylabel("Average Pass Rate")
plt.title("Monthly Progress Messages vs. Average Pass Rate")
plt.grid(True)
plt.show()

# Pearson correlation
corr, pval = pearsonr(df["Weekly_Progress_Messages"], df["Pass_Rate"])
print("Pearson correlation:", round(corr, 3))
print("p-value:", round(pval, 4))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Dosyayı oku
df = pd.read_excel("/content/DSA_Student_List_Final.xlsx", sheet_name="Merged")

# Line plot: Monthly Message Days vs Pass Rate
df_grouped = df.groupby("Weekly_Message_Days")["Pass_Rate"].mean().reset_index()

plt.figure(figsize=(8, 5))
plt.plot(df_grouped["Weekly_Message_Days"], df_grouped["Pass_Rate"], marker='o', linestyle='-')
plt.xlabel("Monthly Message Days")
plt.ylabel("Average Pass Rate")
plt.title("Monthly Message Days vs. Average Pass Rate")
plt.grid(True)
plt.show()

# Pearson correlation
corr, pval = pearsonr(df["Weekly_Message_Days"], df["Pass_Rate"])
print("Pearson correlation:", round(corr, 3))
print("p-value:", round(pval, 4))
```
